**Dataset Description** <br>
There are 395 entries and 33 columns in this data. Attributes for student_data.csv dataset are: 
- school - student's school ('GP' - Gabriel Pereira or 'MS' - Mousinho da Silveira)
- sex - student's sex ('F' - female or 'M' - male)
- age - student's age

- famsize - family size ('LE3' - less or equal to 3 or 'GT3' - greater than 3)
- Pstatus - parent's cohabitation status ('T' - living together or 'A' - apart)
- Medu - mother's education
- Fedu - father's education
- Mjob - mother's job
- Fjob - father's job
- reason - reason to choose this school
- guardian - student's guardian
- traveltime - travel time from home to school
- studytime - weekly study time
- failures - number of past class failures
- schoolsup - extra educational support 
- famsup - family educational support 
- paid - extra paid classes
- activities - extra-curricular activities
- nursery - attended nursery school 
- higher - wants to take higher education
- internet - internet access at home
- romantic - in a romantic relationship
- famrel - quality of family relationships
- freetime - free time after school
- goout - going out with friends 
- Dalc - workday alcohol consumption 
- Walc - weekend alcohol consumption
- health - current health status
- absences - number of school absences
- G1 - first period grade
- G2 - second period grade
- G3 - final grade

**Handling columns with binary values** <br>
Column 'schoolsup', 'famsup', 'paid', 'activities', 'nursery', 'higher', 'internet', 'romantic' have values of yes or no, I want to convert to numerical equiavelent.

```python
def convert_binary(data):
    if data == "yes":
        return 1
    elif data == "no":
        return 0
    
student_df['schoolsup'] = student_df['schoolsup'].apply(convert_binary)
student_df['famsup'] = student_df['famsup'].apply(convert_binary)
student_df['paid'] = student_df['paid'].apply(convert_binary)
student_df['activities'] = student_df['activities'].apply(convert_binary)
student_df['nursery'] = student_df['nursery'].apply(convert_binary)
student_df['higher'] = student_df['higher'].apply(convert_binary)
student_df['internet'] = student_df['internet'].apply(convert_binary)
student_df['romantic'] = student_df['romantic'].apply(convert_binary)
```

```python
# sanity check 
student_df['schoolsup'].value_counts()
```

**Extracting teacher from 'Mjob' and 'Fjob'**<br>
Hypothesis: If the student's Mother or Father works as a teacher, the grade will be higher.

```python
# extracting teacher from the Mjob and Fjob
def extract_teacher(text):
    if "teacher" in text['Mjob'] or "teacher" in text['Fjob'] :
        return 1
    return 0

student_df['parent_is_teacher'] = student_df[['Mjob', 'Fjob']].apply(lambda x: extract_teacher(x),axis=1)
student_df.head()
```

```python
print("Number of student whose parent works as teacher:", len(student_df[student_df['parent_is_teacher']==1]))
```

```python
student_df['average_grade'] = (student_df["G1"] + student_df['G2'] + student_df['G3']) / 3

sns.boxplot(data=student_df, x='parent_is_teacher', y='average_grade')

parent_is_teacher = student_df.query("parent_is_teacher == 1 ")
parent_not_teacher = student_df.query("parent_is_teacher == 0 ")

print(f"""
Mean :
Parent is teacher: {parent_is_teacher.average_grade.mean():.2f}
Parent not teacher: {parent_not_teacher.average_grade.mean():.2f}
""")
```

**Conclusion**<br>
Student whose one of their parent works as a teacher scores slightly higher in test and quizzes.

**Standardization**<br>
Data standardization is the process of rescaling the attributes so that they have mean as 0 and variance as 1.

```python
# importing libraries
from sklearn.preprocessing import StandardScaler

cols = X_train.columns.to_list()

standardizer = StandardScaler()

X_train[cols] = standardizer.fit_transform(X_train[cols])
X_test[cols] = standardizer.transform(X_test[cols])
```

**Training Machine Learning**<br>
R-squared score, mean absolute error (MAE) and mean squared error (MSE) will be used as a performance metrics, as this is a regression model.

```python
# importing libraries
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

**Linear Regression**

```python
linear = LinearRegression()
linear.fit(X_train,y_train)
y_pred = linear.predict(X_test)

# calculate R-squared
r2_linear = r2_score(y_test, y_pred)

# calculate mean absolute error
mae_linear = mean_absolute_error(y_test, y_pred)

# calculate mean squared error
mse_linear = mean_squared_error(y_test, y_pred)

print(f"R2-score: {r2_linear:.4f}")
print(f"MAE score: {mae_linear:.4f}")
print(f"MSE score: {mse_linear:.4f}")
```

Impact of features on the modelling

**k-nearest neighbors (KNN)**

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn = KNeighborsRegressor()
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

# Calculate R-squared
r2_knn = r2_score(y_test, y_pred)

# Calculate mean absolute error
mae_knn = mean_absolute_error(y_test, y_pred)

# calculate mean squared error
mse_knn = mean_squared_error(y_test, y_pred)

print(f"R2-score: {r2_knn:.4f}")
print(f"MAE score: {mae_knn:.4f}")
print(f"MSE score: {mse_knn:.4f}")
```

*Hyper-parameter tuning*

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_neighbors': [3, 5, 7],
    'weights': ['uniform', 'distance'],
    'p': [1, 2]
}

grid_search = GridSearchCV(estimator=knn, param_grid=param_grid, scoring='neg_mean_squared_error', cv=5)
grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
knn_best = grid_search.best_estimator_

y_pred = knn_best.predict(X_test)

# Calculate R-squared
r2_knn_best = r2_score(y_test, y_pred)

# Calculate mean absolute error
mae_knn_best = mean_absolute_error(y_test, y_pred)

# calculate mean squared error
mse_knn_best = mean_squared_error(y_test, y_pred)

print(f"R2-score: {r2_knn_best:.4f}")
print(f"MAE score: {mae_knn_best:.4f}")
print(f"MSE score: {mse_knn_best:.4f}")
```

```python
sns.regplot(x=y_test, y=y_pred).set(title='KNN')
```

**Decision Tree**

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
decision_tree = DecisionTreeRegressor(random_state=0)
decision_tree.fit(X_train, y_train)

y_pred = decision_tree.predict(X_test)

# Calculate R-squared
r2_dt = r2_score(y_test, y_pred)

# Calculate mean absolute error
mae_dt = mean_absolute_error(y_test, y_pred)

# calculate mean squared error
mse_dt = mean_squared_error(y_test, y_pred)

print(f"R2-score: {r2_dt:.4f}")
print(f"MAE score: {mae_dt:.4f}")
print(f"MSE score: {mse_dt:.4f}")
```

*Hyper-parameter tuning*

```python
param_grid = {
    'max_depth': [None, 5, 10, 15],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2']
}

grid_search = GridSearchCV(estimator=decision_tree, param_grid=param_grid, scoring='neg_mean_squared_error', cv=5)
grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
dt_best = grid_search.best_estimator_

y_pred = dt_best.predict(X_test)

# Calculate R-squared
r2_dt_best = r2_score(y_test, y_pred)

# Calculate mean absolute error
mae_dt_best = mean_absolute_error(y_test, y_pred)

# calculate mean squared error
mse_dt_best = mean_squared_error(y_test, y_pred)

print(f"R2-score: {r2_dt_best:.4f}")
print(f"MAE score: {mae_dt_best:.4f}")
print(f"MSE score: {mse_dt_best:.4f}")
```

```python
sns.regplot(x=y_test, y=y_pred).set(title='Decision Tree')
```

**Stochastic Gradient Descent (SGD)**

```python
from sklearn.linear_model import SGDRegressor
```

```python
sgd = SGDRegressor(random_state=0)
sgd.fit(X_train, y_train)

y_pred = sgd.predict(X_test)

# Calculate R-squared
r2_sgd = r2_score(y_test, y_pred)

# Calculate mean absolute error
mae_sgd = mean_absolute_error(y_test, y_pred)

# calculate mean squared error
mse_sgd = mean_squared_error(y_test, y_pred)

print(f"R2-score: {r2_sgd:.4f}")
print(f"MAE score: {mae_sgd:.4f}")
print(f"MSE score: {mse_sgd:.4f}")
```

*Hyper-parameter tuning*

```python
param_grid = {
    'penalty': ['l1', 'l2', 'elasticnet'],
    'alpha': [0.0001, 0.001, 0.01],
    'learning_rate': ['constant', 'optimal', 'invscaling', 'adaptive'],
    'max_iter': [1000, 2000, 5000]
}

sgd = SGDRegressor(random_state=0)

grid_search = GridSearchCV(estimator=sgd, param_grid=param_grid, scoring='neg_mean_squared_error', cv=5)
grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
sgd_best = grid_search.best_estimator_

y_pred = sgd_best.predict(X_test)

# Calculate R-squared
r2_sgd_best = r2_score(y_test, y_pred)

# Calculate mean absolute error
mae_sgd_best = mean_absolute_error(y_test, y_pred)

# calculate mean squared error
mse_sgd_best = mean_squared_error(y_test, y_pred)

print(f"R2-score: {r2_sgd_best:.4f}")
print(f"MAE score: {mae_sgd_best:.4f}")
print(f"MSE score: {mse_sgd_best:.4f}")
```

```python
sns.regplot(x=y_test, y=y_pred).set(title='SGD')
```

**Result of Models**

```python
mae_scores = [mae_linear, mae_knn, mae_knn_best, mae_dt, mae_dt_best, mae_sgd, mae_sgd_best]
r2_scores = [r2_linear, r2_knn, r2_knn_best, r2_dt, r2_dt_best, r2_sgd, r2_sgd_best]
mse_scores = [mse_linear, mse_knn, mse_knn_best, mse_dt, mse_dt_best, mse_sgd, mse_sgd_best]
methods = ["Linear Regression", "KNN", "KNN with Grid Search", "Decision Tree",  "Decision Tree with Grid Search", "SGD", "SGD with Grid Search"]

df_results = pd.DataFrame({
    "R2-Score": r2_scores,
    "MAE": mae_scores,
    "MSE": mse_scores,
},index = methods)

df_results.sort_values(by='R2-Score', ascending=False)
```

From the table above, it can be seen that the Linear Regression model performs the best, which reaches 0.8237 of R2-score, 1.2683 of MAE and 3.5978 of MSE. SGD performs similarly as Linear Regression, followed by Decision Tree at the third place and KNN performs the worst.
